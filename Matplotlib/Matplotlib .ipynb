
$$f(x) = x$$
$$g(x) = x^2$$
$$h(x) = x^3$$


```python
x = np.linspace(0, 4, 1001)


fig, ax = plt.subplots()

ax.grid()
ax.set_title(r"Gráfica simple $\nabla \cdot B = 0$")
ax.plot(x, x, label=r"$f(x) = x$")  #f(x)
ax.plot(x, x**2, label=r"$g(x) = x^2$") # g(x)
ax.plot(x, x**3, label =r"$h(x) = x^3$")
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.text(2, 30, r"$\int_{a} ^{b}$", fontsize=20)
ax.legend()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1001)

f = np.sin(x)
g = np.cos(x)

fig, ax = plt.subplots()

ax.grid()
ax.set_title(r"Funciones trigonométricas")
ax.plot(x, f, color="Red", label=r"$\sin(x)$")
ax.plot(x, g, color="purple", label=r"$\cos(x)$")
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.set_xlim(min(x), max(x))
ax.set_ylim(-1,1)
ax.legend()
```
